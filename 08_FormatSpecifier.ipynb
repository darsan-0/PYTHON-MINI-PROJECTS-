```python
# format specifiers = {:flags) format a value based on what flags are inserted
#.(number)f = round to that many decimal places (fixed point)
#: (number) = allocate that many spaces
#:03 allocate and zero pad that many spaces
# :< = left justify
# :> = right justify
#:^ = center align
#:+ = use a plus sign to indicate positive value
# := = place sign to leftmost position
#: = insert a space before positive numbers
#:, = comma separator
```

```python
price1 = 3.1234
price2 = -423.1234
price3 = 25.1234
```

```python
#.(number)f = round to that many decimal places (fixed point)
print(f"Price1 is ${price1:.2f}")
print(f"Price1 is ${price2:.2f}")
print(f"Price1 is ${price3:.2f}")
```

```python
#: (number) = allocate that many spaces
print(f"Price1 is ${price1:10}")
print(f"Price1 is ${price2:10}")
print(f"Price1 is ${price3:10}")
```

```python
print(f"Price1 is ${price1:010}")
print(f"Price1 is ${price2:010}")
print(f"Price1 is ${price3:010}")
```

```python
# :< = left justify
print(f"Price1 is ${price1:<010}")
print(f"Price1 is ${price2:<010}")
print(f"Price1 is ${price3:<010}")
```

```python
# :> = right justify
print(f"Price1 is ${price1:>010}")
print(f"Price1 is ${price2:>010}")
print(f"Price1 is ${price3:>010}")
```

```python
#:^ = center align
print(f"Price1 is ${price1:^010}")
print(f"Price1 is ${price2:^010}")
print(f"Price1 is ${price3:^010}")
```

```python
#:+ = use a plus sign to indicate positive value
print(f"Price1 is ${price1:+}")
print(f"Price1 is ${price2:+}")
print(f"Price1 is ${price3:+}")
```

```python
p1 = 11111.111
p2 = -22222.1111
p3 = 33333333.1111
```

```python
#:, = comma separator
print(f"Price1 is ${p1:,}")
print(f"Price1 is ${p2:,}")
print(f"Price1 is ${p3:,}")
```

```python
print(f"Price1 is ${p1:+,.2f}")
print(f"Price1 is ${p2:+,.2f}")
print(f"Price1 is ${p3:+,.2f}")
```
